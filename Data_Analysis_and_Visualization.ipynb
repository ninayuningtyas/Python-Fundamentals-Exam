```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Read Data from CSV

```python
#import dataset
df = pd.read_csv('/content/drive/MyDrive/Intensive BC Katalis/movies.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
print("Average of Movie Gross: ", df['gross'].mean())
print("Median of Movie Budget: ", df['budget'].median())
print("Standard Deviation of Movie Score", np.std(df['score']))
```

### Visualization

```python
genre = df['genre']
score = df['score']

fig = plt.figure(figsize=(10,5))

plt.barh(genre, score)
plt.xlabel('Score')
plt.ylabel('Movie Genre')
plt.title('The score of Movie Genre')
plt.show()
```

```python
plt.hist(df['score'])
plt.title('Distribution of Movie Score')
plt.show()
```

```python
plt.scatter(df['budget'],df['gross'])
plt.title("The Distribution Between Budget and Gross")
plt.xlabel("Movie's Budget")
plt.ylabel("Movie's Gross")
```
